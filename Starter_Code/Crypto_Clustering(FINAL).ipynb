```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

### Prepare the Data

```python
# code was written using chatgpt
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler=StandardScaler()
# Fit and transform the data
df_scaled_array = scaler.fit_transform(df_market_data)

# Convert the scaled data back to a DataFrame
df_scaled = pd.DataFrame(df_scaled_array, columns=df_market_data.columns, index=df_market_data.index)

# Check the first few rows of the scaled DataFrame
df_scaled.head()
```

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values=list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia=[]

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    

    kmeans.fit(df_scaled)
    
    # Append the inertia value to the inertia list
    inertia.append(kmeans.inertia_)

# Display the inertia value(code written using chatgpt):
print("Inertia values for different k values:")
for k, inertia_value in zip(k_values, inertia):
    print(f"k={k}: {inertia_value}")
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data= {'k':list(k_values),'inertia':inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df= pd.DataFrame(elbow_data)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

elbow_graph= elbow_df.hvplot.line(
    x='k',
    y='inertia',
    title='Elbow Method for Optimal k',
    xticks=k
)
elbow_graph
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 
--- The best value for k is 4 because it is the point in which the "elbow" forms.

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
original_model=KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled data
original_model.fit(df_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
kprediction=original_model.predict(df_scaled)

# Print the resulting array of cluster values.
kprediction
```

```python
# Create a copy of the DataFrame
df_scaled_predict= df_scaled.copy()
df_scaled_predict.head()
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_market=pca.fit_transform(df_scaled_predict)
# View the first five rows of the DataFrame. 
pca_market[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance=pca.explained_variance_ratio_
print(explained_variance)
print(sum(explained_variance))
```

```python
pca_market_df = pd.DataFrame(
    pca_market,
    columns = ["PCA1", "PCA2", "PCA3"]
)
pca_market_df['coin_id'] = list(df_market_data.index)
pca_market_df=pca_market_df.set_index('coin_id')
pca_market_df.head()
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k=list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia_values = []
# written using chat gpt and course materials
# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for num_clusters in k:
    k_model = KMeans(n_clusters=num_clusters, random_state=1)
    k_model.fit(pca_market)
    inertia_values.append(k_model.inertia_)

# Display the inertia values
print("Inertia values for different k values:")
for num_clusters, inertia in zip(k, inertia_values):
    print(f"k={num_clusters}: {inertia}")
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve2= {"k":k, "inertia":inertia_values}
# Create a DataFrame with the data to plot the Elbow curve
elbow2_df=pd.DataFrame(elbow_curve2)
elbow2_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_graph2=elbow2_df.hvplot.line(
x="k",
y="inertia",
title= "Elbow Curve with PCA",
xticks=k
)
elbow_graph2
```
